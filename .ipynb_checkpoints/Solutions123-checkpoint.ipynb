```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv',low_memory=False)
df.head()
```

```python
df.set_index('report_date', inplace=True)
```

```python
df['value'] = pd.to_numeric(df['value'], errors='coerce')
df['value'].fillna(0)
df['value'] = df['value'].astype('float64') 
```

```python
plt.style.use('ggplot')
```

# Solutions

### 1. Show data field code 'AR0005' values for 'Argentina-Catamarca' for entire duration

```python
t1 = df[(df.location=='Argentina-Catamarca') & (df.data_field_code=='AR0005')].loc[:,'value']
t1=t1.astype(float)
t1.plot(kind='bar', y='value')
plt.title('Cases under study reported in Argentina-Catamarca')
plt.ylabel('No of cases')
```

### 2. Plot 'Argentina-CABA' and 'Argentina-Catamarca' for number of cases values for entire duration to identify the difference.

```python
t1 = df[(df.location=='Argentina-Catamarca') & (df.unit=='cases')].groupby('report_date').sum()['value']
t2 = df[(df.location=='Argentina-CABA') & (df.unit=='cases')].groupby('report_date').sum()['value']
plt.figure(figsize=(15,8))
plt.bar(t1.index,t2,label='Argentina-Catamarca')
plt.bar(t1.index,t1,label='Argentina-CABA')
plt.xticks(rotation='vertical')
plt.ylabel('No of cases')
plt.title('Comparison of the cases recorded in Argentina-Catamarca and Argentina-CABA')
plt.legend()
```

### 3. Plot how data field values are distributed across the locations 'Argentina-Buenos_Aires', 'Argentina-CABA', 'Argentina-Catamarca'

```python
mask = (df.location=='Argentina-Buenos_Aires')|(df.location=='Argentina-CABA')|(df.location=='Argentina-Catamarca')
dft = df[mask].loc[:,['location','data_field','value']]
plt.figure(figsize=(15,30))


lst = ['cumulative_confirmed_imported_cases', 'cumulative_probable_imported_cases', 'cumulative_cases_under_study', 'cumulative_cases_discarded']
for i in range(4):
    tp = pd.DataFrame(dft[dft.data_field == lst[i]].groupby('location').sum()).reset_index()
    tp = tp[tp.value>0]
    ex = np.linspace(0,0.02,tp.shape[0])
    plt.subplot(4,1,i+1)
    plt.pie(tp['value'], labels=tp['location'], autopct='%1.f%%', explode=ex);
    plt.title(lst[i])
```
